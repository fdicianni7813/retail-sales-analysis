# Data Cleaning

This notebook focuses on cleaning and preparing the retail sales dataset for analysis.

```python
import pandas as pd
import numpy as np
```

## Load Raw Dataset

```python
df = pd.read_csv("../data/raw/sales_data.csv")
```

```python
df.head()
```

## Dataset Overview

```python
df.info()
```

```python
df.describe()
```

## Simulate Real-World Data Issues

```python
# Introduce missing values
df.loc[5, "payment_method"] = np.nan
df.loc[12, "stock_remaining"] = np.nan

# Introduce inconsistent category naming
df.loc[20, "category"] = "Cheeses"
df.loc[45, "category"] = "cured meat"

# Duplicate one row
duplicate_row = df.iloc[30]

df = pd.concat([df, duplicate_row.to_frame().T], ignore_index=True)
```

## Check Missing Values

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["category"].unique()
```

## Handle Missing Values

```python
df["payment_method"] = df["payment_method"].fillna("Unknown")

df["stock_remaining"] = df["stock_remaining"].fillna(
    df["stock_remaining"].median()
)
```

## Standardize Category Names

```python
df["category"] = df["category"].replace({
    "Cheeses": "Cheese",
    "cured meat": "Cured Meat"
})
```

## Remove Duplicate Rows

```python
df = df.drop_duplicates()
```

## Verify Cleaned Dataset

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["category"].unique()
```

## Export Cleaned Dataset

```python
df.to_csv("../data/cleaned/cleaned_sales_data.csv", index=False)
```
