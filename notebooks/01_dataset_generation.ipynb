```python
payment_methods = [
    "Credit Card",
    "Cash",
    "PayPal",
    "Debit Card"
]
```

```python
start_date = datetime(2025, 1, 1)
end_date = datetime(2025, 12, 31)
```

## Generate Random Sales Orders

```python
sales_data = []
```

```python
for order_num in range(1, 1001):

    # Randomly select a product
    product = random.choice(list(products.keys()))

    # Extract product information
    product_info = products[product]

    category = product_info["category"]
    price = product_info["price"]
    product_cost = product_info["cost"]

    # Generate quantity sold
    quantity_sold = random.randint(1, 5)

    # Calculate profit
    profit = round((price - product_cost) * quantity_sold, 2)

    # Simulate remaining stock
    stock_remaining = random.randint(0, 150)

    # Random payment method
    payment_method = random.choice(payment_methods)

    # Generate random order date
    random_days = random.randint(0, 364)
    order_date = start_date + timedelta(days=random_days)

    # Append order data
    sales_data.append({
        "order_id": f"ORD{order_num:04d}",
        "order_date": order_date,
        "customer_id": f"CUST{random.randint(100,999)}",
        "product_name": product,
        "category": category,
        "price": price,
        "quantity_sold": quantity_sold,
        "product_cost": product_cost,
        "profit": profit,
        "stock_remaining": stock_remaining,
        "payment_method": payment_method
    })
```

## Create DataFrame and Export CSV

```python
df = pd.DataFrame(sales_data)
```
